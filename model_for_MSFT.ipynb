# dataframe dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('DataFrame.csv')
data.head()
```

```python
data.drop('Unnamed: 7' , axis = 1 , inplace = True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
import datetime
data['Date'] = pd.to_datetime(data['Date'] , format = "%Y%m%d")
```

```python
data['Day']=pd.to_datetime(data['Date'], format="%Y/%m/%d").dt.day
data['Month']=pd.to_datetime(data['Date'], format="%Y/%m/%d").dt.month
data['Year']=pd.to_datetime(data['Date'], format="%Y/%m/%d").dt.year
data.drop(['Date'],axis=1,inplace=True)
```

```python
data['Hour']=pd.to_datetime(data['Time']).dt.hour
data['Minute']=pd.to_datetime(data['Time']).dt.minute
data.drop(['Time'],axis=1,inplace=True)
```

```python
data.drop(['Type'],axis=1,inplace=True)
data.head()
```

```python
data.describe()   #Statistics of the Dataset
```

```python
data.info()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=data[["open","low","high","close"]])
```

```python
color = sns.PairGrid(data)
color.map_upper(sns.scatterplot,color = 'aqua')
color.map_lower(sns.scatterplot,color = 'pink')
color.map_diag(plt.hist)
```

# msft dataset

```python
df2 = pd.read_csv('MSFT.csv')
df2.head()
```

```python
df2['Date'] = pd.to_datetime(df2['Date'] , format = "%Y-%m-%d")
```

```python
df2.isnull().sum()
```

```python
df2.duplicated().sum()
```

```python
df2.describe()
```

```python
df2.info()
```

```python
df2 = df2.set_index('Date')
df2
```

```python
sns.pairplot(data = df2)  
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df2[["Open","High","Low","Close","Adj Close"]])
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df2[["Volume"]])
```

```python
#removing outliners
```

```python
iqropen= df2['Open'].quantile(0.75) - df2['Open'].quantile(0.25) 
iqropen
```

```python
uopen = df2['Open'].quantile(0.75)+(iqropen*1.5)
lopen = df2['Open'].quantile(0.25)-(iqropen*1.5)
print(uopen, lopen)
```

```python
iqrhigh = df2['High'].quantile(0.75) - df2['High'].quantile(0.25) 
iqrhigh
```

```python
uhigh = df2['High'].quantile(0.75)+(iqrhigh*1.5)
lhigh = df2['High'].quantile(0.25)-(iqrhigh*1.5)
print(uhigh,lhigh)
```

```python
iqrclose = df2['Close'].quantile(0.75) - df2['Close'].quantile(0.25) 
iqrclose
```

```python
uclose= df2['Close'].quantile(0.75)+(iqrclose*1.5)
lclose = df2['Close'].quantile(0.25)-(iqrclose*1.5)
print(uclose,lclose)
```

```python
iqrlow = df2['Low'].quantile(0.75) - df2['Low'].quantile(0.25) 
iqrlow
```

```python
ulow = df2['Low'].quantile(0.75)+(iqrlow*1.5)
llow = df2['Low'].quantile(0.25)-(iqrlow*1.5)
print(ulow,llow)
```

```python
iqradjclose = df2['Adj Close'].quantile(0.75) - df2['Adj Close'].quantile(0.25) 
iqradjclose
```

```python
uadjclose = df2['Adj Close'].quantile(0.75)+(iqradjclose*1.5)
ladjclose = df2['Adj Close'].quantile(0.25)-(iqradjclose*1.5)
print(uadjclose,ladjclose)
```

```python
iqrvolume = df2['Volume'].quantile(0.75) - df2['Volume'].quantile(0.25) 
iqrvolume
```

```python
uvolume = df2['Volume'].quantile(0.75)+(iqrvolume*1.5)
lvolume = df2['Volume'].quantile(0.25)-(iqrvolume*1.5)
print(uvolume, lvolume)
```

```python
df2.loc[df2['Open'] >= uopen , 'Open'] = uopen
df2.loc[df2['Close'] >= uclose, 'Close'] = uclose
df2.loc[df2['High'] >= uhigh , 'High'] = uhigh
df2.loc[df2['Low'] >= ulow , 'Low'] = ulow
df2.loc[df2['Adj Close'] >= uadjclose , 'Adj Close'] = uadjclose
df2.loc[df2['Volume'] >= uvolume , 'Volume'] = uvolume
```

```python
df2
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df2[["Open","High","Low","Close","Adj Close"]])
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df2[["Volume"]])
```

# Modelling dataframe dataset

```python
data.head()
```

```python
X = data.loc[:,data.columns!='close']
X
```

```python
y=data['close']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
X_train_std=sc.fit_transform(X_train)
```

```python
X_test_std=sc.transform(X_test)
```

```python
X_train_std
```

# svm linear on dataframe dataset

```python
from sklearn.svm import SVR
dataframelinear = SVR(kernel='linear', C = 22)
```

```python
dataframelinear.fit(X_train_std,y_train)
```

```python
y_train_pred = dataframelinear.predict(X_train_std)
```

```python
y_test_pred = dataframelinear.predict(X_test_std)
```

```python
result = pd.DataFrame({'Actual' : y_test , 'Predicted' : y_test_pred}).reset_index()[['Actual' , 'Predicted']]
result
```

```python
plt.plot(y_test,result['Predicted'],color='red')
plt.plot(y_test,result['Actual'],color='blue')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test,y_test_pred)
print("Mean_squared_error is: ", mse)
```

```python
r2score = r2_score(y_test , y_test_pred)
print("R2 score is: " , r2score)
```
